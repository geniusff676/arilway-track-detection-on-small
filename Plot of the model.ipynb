```python
import json
import matplotlib.pyplot as plt

# Path to the metrics.json file
file_path = 'output/metrics.json'

# Lists to store iteration numbers and total loss values
iterations = []
total_loss = []

# Read the JSON lines from the file
with open(file_path, 'r') as f:
    for line in f:
        data = json.loads(line)
        if 'iteration' in data and 'total_loss' in data:
            iterations.append(data['iteration'])
            total_loss.append(data['total_loss'])

# Plot the total loss over iterations
plt.figure(figsize=(7, 6))
plt.plot(iterations, total_loss, label='Total Loss  ',color ='red')
plt.xlabel('Iteration')
plt.ylabel('Total Loss')
plt.title('Total Loss over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (replace these with your actual AP values per category)
categories = ["mAP50 ", "mAP75", "mAP"]
ap_values = [81.8543373519633, 56.51068510714777,49.39979304457765]  # Replace 'None' with 'nan'

# Handle NaN values
ap_values = [val if val is not None else 0 for val in ap_values]

plt.figure(figsize=(7, 7))
bars = plt.bar(categories, ap_values, color='brown')
for bar, val in zip(bars, ap_values):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{val:.2f}%', ha='center', va='bottom')

plt.xlabel('Category')
plt.ylabel('Average Precision (AP)')
plt.title('Average Precision (AP) for Each Category')
plt.ylim(0, 100)  # Assuming AP is a percentage, otherwise adjust the scale
plt.show()
```
